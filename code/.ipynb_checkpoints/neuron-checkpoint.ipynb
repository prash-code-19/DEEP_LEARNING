```python
import torch
import torch.nn as nn
```

```python
# input
torch.manual_seed(42)
inputs = torch.tensor([1.0,2.0,3.0])
```

```python
neuron = nn.Linear(in_features =3, out_features=1,)
```

```python
output = neuron(inputs)
```

```python
output
```

```python
neuron.weight
```

```python
neuron.bias
```

```python
# verify the formula

neuron.weight @ inputs + neuron.bias
```
